```python
#iniciando o menu de opções
menu = '''

[1] Depositar
[2] Sacar
[3] Extrato
[0] Sair

=> '''

#declarando as variáveis
saldo = 0
limite = 500
extrato = ''
numero_saques = 0
LIMITE_SAQUES = 3

#formando o sistema com loop while
while True:
    opcao = input(menu)
    if opcao == '1':
        valor = float(input('Informe o valor do seu depósito: '))#perguntando ao usuario o valor que deseja depositar.

        if valor > 0:#realizando a verificacao se o deposito é negativo ou positivo.

            saldo += valor#sendo positivo, adiciona o valor a variavel saldo.

            extrato += f'Depósito: R${valor:.2f}\n'#concatenando a variavel extrato com o valor depositado com 2 casas decimais e adc direto ao extrato.

        else:
            print('Operação inválida, digite um valor positivo.')#mensagem para caso o valor seja negativo.



    elif opcao == '2':
        valor = float(input('Informe o valor do saque: '))#questionando o valor que o usuario deseja sacar.

        excedeu_saldo = valor > saldo#verificando se excedeu o saldo.

        excedeu_limite = valor > limite#verificando se excedeu o limite.

        excedeu_saques = numero_saques >= LIMITE_SAQUES#verificando se excedeu o limite de saques.

        if excedeu_saldo:
            print('A operação falhou, seu saldo é insuficiente.')#mensagem de falha se o valor for maior que o saldo.

        elif excedeu_limite:
            print('A operação falhou, o valor do saque exede o limite.')#mensagem de falha se o valor for maior que o limite.

        elif excedeu_saques:
            print('A operação falhou, seu limite de saques diários foi atingido.')#mensagem de falha se o valor for maior que o limite.

        elif valor > 0:#verificando se o valor solicitado no saque e maior que 0.

            saldo -= valor#retirando o valor do saque do saldo.

            extrato += f'Saque: R$ {valor:.2f}\n'#concatenando dentro do extrato a operação do saque.

            numero_saques += 1#incrementando o numero de saques.

        else:
            print('A operação falhou, o valor informado é inválido.')#informando que o valor deve ser maior que 0.



    elif opcao == '3':
        print('\n========== Extrato ==========')#inserindo um cabeçalho.
        print('Não realizou nenhuma movimentação' if not extrato else extrato)#verificando se o extrato está "vazio" e dando informativo.
        print(f'\nSaldo final: R$ {saldo:.2f}')#exibindo o saldo final da conta.
        print('=============================')#inserindo um rodapé.
        #as opções do "\n" servem para quebrar uma linha.


    elif opcao == '0':#opção para sair do "sistema"
        break

    else:
        print('Opção inválida, digite uma volor que pertence ao menu.')#mensagem de erro caso o numero selecionado não pertença ao menu.
```
